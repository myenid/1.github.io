```python
import matplotlib.pyplot as plt
```

```python
data.INDUS.plot(kind = 'line', color = 'r', label = 'INDUS', linewidth=1, grid = True)
data.AGE.plot(kind = 'line', color = 'g', label = 'AGE', linewidth=1, grid = True)

plt.legend()
plt.title("Line plot of INDUS & AGE")
plt.xlabel('x: Range Index')
plt.ylabel('y: INDUS & AGE value')
plt.show()
```

```python
data.RM.plot(kind = 'hist', bins = 100)
plt.title("Hist of RM")
plt.xlabel("x:RM")
plt.ylabel("y:value")

plt.show()
```

```python
data.RM.plot(kind = 'hist', bins = 100, cumulative = True)
plt.title("Cumulative hist of RM")
plt.xlabel("x: RM")
plt.ylabel("y: Value")

plt.show()
```
