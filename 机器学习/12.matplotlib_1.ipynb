matplotlib图形工具，曲线，柱状图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.random.random(5)
print(y)
```

```python
x = range(5)
```

```python
plt.plot(x, y)
```

```python
y1 = np.random.random(10)
y2 = y1 + 3
print(y1)
print(y2)
```

```python
x = range(10)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.arange(10)
print(x)
```

```python
y1 = x + 3000
y2 = x**4
y3 = np.exp(x)
```

```python
plt.plot(x, y1, linewidth = 1)
plt.plot(x, y2, linewidth = 4)
plt.plot(x, y3, linewidth = 8)
```

```python
plt.plot(x, y1, 'r', linewidth = 1)
plt.plot(x, y2, 'y', linewidth = 2)
plt.plot(x, y3, 'b', linewidth = 4)
```

```python
plt.plot(x, y1, 'ro--', linewidth = 1)
plt.plot(x, y2, 'y^-', linewidth = 2)
plt.plot(x, y3, 'b-', linewidth = 4)
```

```python
plt.plot(x, y1, 's', linewidth = 1)
plt.plot(x, y2, '--', linewidth = 2)
plt.plot(x, y3, '*', linewidth = 4)
plt.show()
```
