在一个图表中绘制多个子图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
y1 = x + 2
y2 = x**2
y3 = np.log(2*x + 1)
y4 = (1/(x+1))*np.sin(x)
```

```python
#subplot(arg1, arg2, arg3), arg1:代表子图的行数，arg2：代表子图的列数，arg3：当前子图的编号
plt.subplot(2, 2, 1)  #plt.subplot(221)
plt.plot(x, y1, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'y')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'b')

plt.show()
```
